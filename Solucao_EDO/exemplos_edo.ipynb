### 2.1 Movimento Horizontal de um carrinho



#### Dados para simula√ß√£o

```python
m = 1.0 # Massa da p√°rticula [kg]
b = 0.1 # Coeficiente de atrito [kg/s]
h = 0.01 # Passo de tempo [s]
```

#### Fun√ß√£o

```python
dy = lambda t,v: -b*v/m
f = lambda t: 10*math.exp(-b*t/m) # Solu√ß√£o anal√≠tica
```

#### Condi√ß√µes Iniciais

```python
ti = 0; tf = 60; 
n = int((tf-ti)/h)
t = 0; v = 10.0
```

```python
x_plot=[]; y_RK4=[]; y_analytical=[]

for i in range(1, n+1):
    x_plot.append(t); y_RK4.append(v); y_analytical.append(f(t))
    k1 = dy(t,v)
    k2 = dy(t+h/2, v+k1*h/2)
    k3 = dy(t+h/2, v+k2*h/2)
    k4 = dy(t+h, v+k3*h)
    
    v = v + h/6*(k1 + 2*k2 + 2*k3 + k4)
    
    t = t + h

plt.figure(figsize=(10, 5))
x_plot.append(t); y_RK4.append(v); y_analytical.append(f(t))
plt.plot(x_plot,y_RK4,'ro',x_plot,y_analytical)
plt.xlabel('x'); plt.ylabel('v')
plt.legend(['RK4','Analytical'])
plt.grid()
```

### 2.2 Velocidade terminal de uma gota



#### Dados para simula√ß√£o

```python
D = 1.5*10**(-6) # Di√¢metro da gota de √≥leo [m]
rho = 840.0 # Densidade do l√≠quido [kg/m¬≥]
g = 9.81 # Acelera√ß√£o da gravidade [m/s¬≤]
V = math.pi*D**3/6 # Volume da gota [m¬≥]
m = rho*V # Massa da part√≠cula [kg]
beta = 1.6*10**(-4) # Viscosidade din√¢mica do ar [kg/(ms)]
b = beta*D # Coeficiente de atrito linear [kg/s]
h = 1*10**(-7) # Passo de tempo [s]
```

#### Fun√ß√£o

```python
dy = lambda t,v: -b*v/m + g
f = lambda t: 0*math.exp(-b*t/m) + (m*g/b)*(1 - math.exp(-b*t/m)) # Solu√ß√£o anal√≠tica
```

#### Condi√ß√µes Iniciais

```python
ti = 0; tf = 5*10**(-5); 
n = int((tf-ti)/h)
t = 0; v = 0;
```

```python
x_plot=[]; y_RK4=[]; y_analytical=[]

for i in range(1, n+1):
    x_plot.append(t); y_RK4.append(v); y_analytical.append(f(t))
    k1 = dy(t,v)
    k2 = dy(t+h/2, v+k1*h/2)
    k3 = dy(t+h/2, v+k2*h/2)
    k4 = dy(t+h, v+k3*h)
    
    v = v + h/6*(k1 + 2*k2 + 2*k3 + k4)
    
    t = t + h
    
plt.figure(figsize=(10, 5))
x_plot.append(t); y_RK4.append(v); y_analytical.append(f(t))
plt.plot(x_plot,y_RK4,'ro',x_plot,y_analytical)
plt.xlabel('x'); plt.ylabel('v')
plt.legend(['RK4','Analytical'])
plt.grid()
plt.show()
```

### 2.4 Sistema massa-mola


#### Dados para simula√ß√£o

```python
m = 1.0 # Massa da p√°rticula [kg]
k = 0.1 # Coeficiente [kg/s¬≤]
w = math.sqrt(k/m) # Frequ√™ncia [Hz]
h = 0.01 # Passo de tempo [s]

A1 = -10/w; j=0
A2 = 10/w**2;
```

### Fun√ß√£o

```python
dv = lambda t,x: -k*x/m
dx = lambda t,v: v
f = lambda t: -w*A1*math.cos(w*t+j) # Solu√ß√£o anal√≠tica
fx = lambda t: w*A2*math.sin(w*t+j) # Solu√ß√£o anal√≠tica
```

#### Condi√ß√µes Iniciais

```python
ti = 0; tf = 60; 
n = int((tf-ti)/h)
t = 0; v = 10.0; x=0
```

```python
t_plot=[]; v_RK4=[]; v_analytical=[]; x_RK4=[]; x_analytical=[]

for i in range(1, n+1):
    t_plot.append(t); v_RK4.append(v); v_analytical.append(f(t))
    x_RK4.append(x); x_analytical.append(fx(t))
    
    k1x = dx(t,v)
    k2x = dx(t+h/2, v+k1x*h/2)
    k3x = dx(t+h/2, v+k2x*h/2)
    k4x = dx(t+h, v+k3x*h)
    
    x = x + h/6*(k1x + 2*k2x + 2*k3x + k4x)
    
    k1 = dv(t,x)
    k2 = dv(t+h/2, x+k1*h/2)
    k3 = dv(t+h/2, x+k2*h/2)
    k4 = dv(t+h, x+k3*h)
    
    v = v + h/6*(k1 + 2*k2 + 2*k3 + k4)
    
    t = t + h

plt.figure(figsize=(10, 5))
t_plot.append(t); v_RK4.append(v); v_analytical.append(f(t))
plt.plot(t_plot,v_RK4,'ro',t_plot,v_analytical)
plt.xlabel('t'); plt.ylabel('v')
plt.legend(['RK4','Analytical'])
plt.grid()

plt.figure(figsize=(10, 5))
x_RK4.append(x); x_analytical.append(fx(t))
plt.plot(t_plot,x_RK4,'ro',t_plot,x_analytical)
plt.xlabel('t'); plt.ylabel('x')
plt.legend(['RK4','Analytical'])
plt.grid()

plt.figure(figsize=(10, 5))
#x_RK4.append(x); x_analytical.append(fx(t))
plt.plot(x_RK4,v_RK4,'ro',x_analytical,v_analytical)
plt.xlabel('x'); plt.ylabel('v')
plt.legend(['RK4','Analytical'])
plt.grid()
```

### 2.5 Sistema massa-mola vertical


#### Dados para simula√ß√£o

```python
m = 0.15 # Massa da p√°rticula [kg]
k = 0.1 # Coeficiente [kg/s¬≤]
h = 0.1 # Passo de tempo [s]
g = 9.81 # Acelera√ß√£o da gravidade [m/s¬≤]
D = 7*10**(-2) # Di√¢metro da part√≠cula [m]
```

### Fun√ß√£o

```python
dv = lambda t,y: -k*y/m + g
dy = lambda t,v: v
```

#### Condi√ß√µes Iniciais

```python
ti = 0; tf = 30; 
n = int((tf-ti)/h)
t = 0; v = 10.0; y=0
```

```python
t_plot=[]; v_RK4=[]; y_RK4=[];

for i in range(1, n+1):
    t_plot.append(t); v_RK4.append(v); y_RK4.append(y);
    
    k1y = dy(t,v)
    k2y = dy(t+h/2, v+k1y*h/2)
    k3y = dy(t+h/2, v+k2y*h/2)
    k4y = dy(t+h, v+k3y*h)
    
    y = y + h/6*(k1y + 2*k2y + 2*k3y + k4y)
    
    k1 = dv(t,y)
    k2 = dv(t+h/2, y+k1*h/2)
    k3 = dv(t+h/2, y+k2*h/2)
    k4 = dv(t+h, y+k3*h)
    
    v = v + h/6*(k1 + 2*k2 + 2*k3 + k4)
    
    t = t + h

plt.figure(figsize=(10, 5))
t_plot.append(t); v_RK4.append(v);
plt.plot(t_plot,v_RK4,'k')
plt.xlabel('t'); plt.ylabel('v')
plt.legend(['RK4'])
plt.grid()

plt.figure(figsize=(10, 5))
y_RK4.append(y)
plt.plot(t_plot,y_RK4,'k')
plt.xlabel('t'); plt.ylabel('y')
plt.legend(['RK4'])
plt.grid()

plt.figure(figsize=(10, 5))
plt.plot(y_RK4,v_RK4,'k')
plt.xlabel('y'); plt.ylabel('v')
plt.legend(['RK4','Analytical'])
plt.grid()
```
